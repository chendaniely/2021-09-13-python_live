```python
billboard = pd.read_csv("../data/billboard.csv")
```

```python
billboard
```

```python
billboard_tidy = billboard.melt(id_vars=["year", "artist", "track", "time", "date.entered"],
              value_name="rank",
              var_name="week")
```
